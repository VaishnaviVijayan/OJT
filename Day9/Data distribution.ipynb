```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Example dataset
data = np.random.normal(loc=0, scale=1, size=1000)
df = pd.DataFrame(data, columns=['values'])

# Create histogram
df['values'].hist(bins=30, edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
